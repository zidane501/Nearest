```python
import pytesseract
import cv2
# img = cv2.imread('./img/page0.jpg')
img = cv2.imread('./blackWhitePrice.jpg')
w,h = 540,900
img = cv2.resize(img, (w,h))
hImg, wImg, _ = img.shape

boxes = pytesseract.image_to_boxes(img)
print(boxes)

for b in boxes.splitlines():
  b = b.split(' ')
  x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
  cv2.rectangle(img, (x, hImg - y), (w, hImg - h), (50, 50, 255), 1)
  cv2.putText(img, b[0], (x, hImg - y + 13), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (50, 205, 50), 1)

cv2.imshow('Detected text', img)
cv2.waitKey(0)
```

```python
import numpy as np
import imutils
import cv2

img_path = './img/page0.jpg'

# construct the argument parse and parse the arguments
# load the image
image = cv2.imread(img_path)
w,h = 540,900
image = cv2.resize(image, (w,h))

# Finding Shapes in Images using Python and OpenCV
# find all the 'black' shapes in the image
lower = np.array([255-15, 255-15, 255-15])
upper = np.array([255, 255, 255])
shapeMask = cv2.inRange(image, lower, upper)

# find the contours in the mask
cnts = cv2.findContours(shapeMask.copy(), cv2.RETR_EXTERNAL,
	cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
print("I found {} white shapes".format(len(cnts)))
print(cnts[0])
cv2.imshow("Mask", shapeMask)
# loop over the contours
for c in cnts:
	if len(c)>150:
		cv2.drawContours(image, [c], -1, (0, 255, 0), 2)
		cv2.imshow("Image", image)
		# cv2.waitKey(0)

cv2.imwrite("blackWhitePrice.jpg", shapeMask)
```
