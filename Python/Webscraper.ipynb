```python
def create_driver(url):

        option = webdriver.ChromeOptions()
        option.add_argument('headless')

        driver = webdriver.Chrome('chromedriver', options=option)  # Optional argument, if not specified will search path.
        # driver = webdriver.Chrome('chromedriver')  # Optional argument, if not specified will search path.

        driver.set_window_position(0, 0)
        driver.set_window_size(2048, 2048)

        driver.get(url)
        return driver
```

```python

from dotenv import load_dotenv
load_dotenv()

GOOGLE_MAPS_APIKEY2 = os.getenv('GOOGLE_MAPS_APIKEY2')

def extract_lat_long_via_address(address_or_zipcode):
    
    address_or_zipcode = address_or_zipcode.split()
    address_or_zipcode = '+'.join(address_or_zipcode,)
    lat, lng = None, None
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    endpoint = f"{base_url}?address={address_or_zipcode}&key={GOOGLE_MAPS_APIKEY2}"
    # see how our endpoint includes our API key? Yes this is yet another reason to restrict the key
    r = requests.get(endpoint)
    if r.status_code not in range(200, 299):
        return None, None
    else:
        print("r.json()['results']",r.json()['results'])
    try:
        '''
        This try block incase any of our inputs are invalid. This is done instead
        of actually writing out handlers for all kinds of responses.
        '''
        results = r.json()['results'][0]
        lat = results['geometry']['location']['lat']
        lng = results['geometry']['location']['lng']
        print(lat, lng)
    except:
        print("except")
        pass
    return {'lat': lat, 'lng': lng}
extract_lat_long_via_address("Hans Muhles Gade 16")
```
